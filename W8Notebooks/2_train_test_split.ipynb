# Train-Test Split

In our introductory demonstration of machine learning using the diabetes dataset, we set aside the last five examples as a test set. We cannot normally select our test examples from the end of a dataset. It will frequently be the case that the examples appear in the dataset in some order such that the end examples are not a representative sample. We have already seen this. The 150 examples in the iris dataset are grouped by species so if we selected the last 5 examples for testing, all the test examples would be species virginica. Clearly we could not perform a valid test of our machine learning with such test set. 

Scikit-learn provides an easy to use tool for performing a train-test split using a random sample from the dataset. 
Unsurprisingly, the tool is called [train_test_split](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html). 

It is easy to use. Nevertheless, it is useful to review importing and manipulating data, so we demonstrate it both on the Diabetes dataset and the iris dataset. 

---

### Import libraries 

We first import `train_test_split` as shown in the cell below.

```python
# import libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn import datasets
from sklearn.model_selection import train_test_split
```

## Example: Diabetes dataset

We begin by loading the dataset and printing the shapes of `X` and `y`. (It is always a good idea to print out some basic information after importing data, whether with pandas, or as here, using scilik-learn. Pandas Dataframes offer many more possibilities, but one can always print the shapes of the data.)


```python
# Load the diabetes dataset
X, y = datasets.load_diabetes(return_X_y=True)

# Print the shapes of X and y to confirm
print(X.shape)
print(y.shape)
```

We confirm that the design matrix `X` has 442 row (samples) and 10 features (columns), and that the target `y` also has length 442.

---

We now use `train_test_split` to split our samples. We use 


```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

**Exercise:** In the cell below print the shapes of `X_train`, `y_train`, `X_test` and `y_test`. 

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

---

You see that far more that 5 examples have been set aside for testing. What percentage of the total samples have been set aside for testing?

The cell below makes a scatter plot showing the training and testing samples. 

```python
plt.scatter(X_train[:,2], X_train[:,3])
plt.scatter(X_test[:,2], X_test[:,3])
plt.show()
```

We choose to show column (feature) indices 2 and 3 only because these are what we considered in the first notebook.

### test_size and random_state parameters

`train_test_split` takes a parameter `test_size` as illustrated below.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
plt.scatter(X_train[:,2], X_train[:,3])
plt.scatter(X_test[:,2], X_test[:,3])
plt.show()
```
